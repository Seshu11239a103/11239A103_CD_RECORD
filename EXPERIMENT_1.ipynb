1.E-commerce Discount Rule


```python
import re

def check_discount_rule(s):
    pattern = r"^DISC-\d{4}-([1-9][0-9])$"
    if re.fullmatch(pattern, s):
        return "Valid Discount Code"
    else:
        return "Invalid Discount Code"

# Test
test_strings = ["DISC-1234-20", "DISC-9999-05", "DISC-12A4-30"]

for t in test_strings:
    print(t, "->", check_discount_rule(t))
```

2. Car Parking Ticket Forma

```python
import re

def check_parking_ticket(s):
    pattern = r"^TN-[A-Z]{2}-\d{4}$"
    if re.fullmatch(pattern, s):
        return "Valid Parking Ticket"
    else:
        return "Invalid Parking Ticket"

# Test
test_tickets = ["TN-AB-1234", "TN-A1-5678", "TN-XYZ-9999"]

for t in test_tickets:
    print(t, "->", check_parking_ticket(t))
```
