a) Online Food Ordering App – Keyword Check

```python
def food_app_keyword(word):
    keywords = {"order", "cancel", "menu", "cart", "pay"}

    if word.lower() in keywords:
        print("Keyword in Food App")
    else:
        print("Not a Keyword in Food App")

# Input
food_app_keyword(input("Enter word: "))
```

b) Library Management System – Keyword Check

```python
def library_keyword(word):
    keywords = {"issue", "return", "book", "fine", "catalog"}

    if word.lower() in keywords:
        print("Keyword in Library System")
    else:
        print("Not a Keyword in Library System")

# Input
library_keyword(input("Enter word: "))
```
