a) Online Shopping Price Calculation

```python
input_str = input("Enter expression: ") + "$"
stack = ""

print("Stack\tInput\tAction")

while input_str:
    stack += input_str[0]
    input_str = input_str[1:]
    print(stack, "\t", input_str, "\tShift")

    if stack in ["id", "E+E", "E*E"]:
        stack = "E"
        print(stack, "\t", input_str, "\tReduce")

if stack == "E$":
    print("Accepted")
else:
    print("Rejected")
```
