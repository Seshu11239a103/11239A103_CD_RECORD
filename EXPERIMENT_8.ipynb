a) Online Shopping Expression

```python
def shift_reduce(expr):
    stack = []
    tokens = expr.split()

    print("\nParsing Steps:")
    for t in tokens:
        stack.append(t)
        print("Shift:", stack)

        while True:
            if len(stack) >= 3 and stack[-1] == "id" and stack[-3] == "id":
                if stack[-2] in ['+', '*']:
                    stack = stack[:-3] + ["E"]
                    print("Reduce to E:", stack)
            else:
                break

    if stack == ["E"]:
        print("Accepted (Valid Expression)")
    else:
        print("Rejected")


expr = input("Enter Online Shopping Expression (use id + id): ")
shift_reduce(expr)
```
